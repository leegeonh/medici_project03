### 2. 수집 방법





import statsapi  

statsapi.get('team', {'teamId':110}) # 구장명을 알기 위한 데이터  
 
statsapi.standings_data(leagueId=103,date='10/05/2022') # 시즌별 팀의 랭킹을 알기 위한 코드  
  
statsapi.get('sports_players',{'season':2022}) # 선수의 개인 정보를 알기 위한 코드  

statsapi.player_stat_data(Baltimore Orioles, group="pitching", type="YearByYear") # 투수의 연간 기록을 알기 위한 코드  
statsapi.player_stat_data(Baltimore Orioles, group="hitting", type="YearByYear") # 타자의 연간 기록을 알기 위한 코드  

statsapi.boxscore_data(game_number)['homePitchers'] # 각 선수별 게임별 데이터를 알기 위한 코드
