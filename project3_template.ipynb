# [특정 팀에서의 이적이 개인 성적에 미치는 영향]

*   기간: 2023-05-24 ~ 2023-06-01
*   리더: 이건희
*   정의: 팀과 이적효과에 대한 관계성 분석
*   목적: KBO에는 '탈쥐효과'라는 하나의 밈이 있는데 이는 'LG twins를 나가 다른 팀으로 이적하면 더 좋은 성적을 낸다'는 의미이다.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;이적하는 것이 성적향상에 영향을 끼치는지에 대하여 mlb 데이터를 이용하여 알아보고  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;나아가 성적에 영향을 미치는 요인에는 어떤 요인들이 있는지 알아보고자 한다.

● 이적하는 것이 성적에 영향을 미치는지 알아보기 위해 이적여부와 성적변화의 상관관계  
&nbsp;&ensp;(구분 : 타자 / 선발투수 / 중간계투 및 마무리투수)

● 우리 팀이 다른 팀에 비해 부족한 부분을 알기 위해 2022시즌을 기준으로 전체 팀별 데이터를 확인
