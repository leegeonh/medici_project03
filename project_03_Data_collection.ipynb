```python
def player_record (year, player, player_list):

    for player_id, position_code in player:
        exist = False

        for year_player_list in player_list:
            for id in year_player_list:
                if f'{player_id}.json' == id[2:]:
                    exist = True
                    break        
        
        if exist == False:
            time.sleep(1) 
            if position_code == '1':
                pd.DataFrame(statsapi.player_stat_data(player_id, group="pitching", type="YearByYear")).to_json(f'./Data/player_record/{year}/p_{player_id}.json')
            elif position_code == 'Y':
                pd.DataFrame(statsapi.player_stat_data(player_id, group="pitching", type="YearByYear")).to_json(f'./Data/player_record/{year}/p_{player_id}.json')
                pd.DataFrame(statsapi.player_stat_data(player_id, group="hitting", type="YearByYear")).to_json(f'./Data/player_record/{year}/b_{player_id}.json')
            else:
                pd.DataFrame(statsapi.player_stat_data(player_id, group="hitting", type="YearByYear")).to_json(f'./Data/player_record/{year}/b_{player_id}.json')

        else:
            continue
```

```python
for year in range(2021, 2012, -1):
    temp = pd.read_json(f'./Data/player/player_{year}.json')
    temp = temp.transpose()
    player = list(zip(temp['id'], pd.json_normalize(temp['primaryPosition'])['code']))

    exist_list =list()
    
    for file in os.walk(f'./Data/player_record'):
        exist_list.append(file[2])
        
    player_record(year, player, exist_list)
```
